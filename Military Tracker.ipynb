In 1922, the Ukrainian Soviet Socialist Republic joined the Russian Federal Soviet Socialist Republic, the Byelorussian Soviet Socialist Republic and the Transcaucasian Soviet Socialist Republic, to create the Union of Soviet Socialist Republics, what we know as the Soviet Union, or the USSR. The Ukraine, as it was known back then, was an autonomous entity, yet it was subservient to Moscow on matters of defense, economic planning and foreign policy.

In 1954, a year after he came to power after Joe Stalin's death, Nikita Khrushchev handed over the Crimea to the Ukrainian SSR. Historians speculate his decision to do so was motivated by the suffering the Ukraine endured during the catastrophic famine in the early 1930s, that was entirely Stalin's handiwork. However, most Crimeans didn't like the decision, and of course, were never consulted, or informed.

In 1991, after the USSR broke up, Ukraine announced itself as a sovereign republic. The Crimea was given a fair degree of autonomy within the republic. 

But there were complications. Ukraine possesed the second-largest stockpile of nuclear weapons, after the newly-formed Russian Federation. That  was a direct contravention to the Nuclear Non-proliferation Treaty - so in 1994, the US, the UK, Russia and Ukraine signed the Budapest Memorandum. It was an agreement where Ukraine agreed to give up its nuclear arsenal, in exchange for iron-clad guarantees of territorial integrity, by all the signatories. Obviously, this was violated by Russia, when the Duma voted to send in troops into Ukraine.

In the last week of February 2014, President Yanukovych was forced out of office, for his brutal response to protests demanding greater integration with the EU. An interim government was installed, with the mandate to solve Ukraine's current debt crisis, and to work towards greater integration with the EU and NATO.

This was unacceptable to the Kremlin. 20-plus years after the fall of the USSR, Russia still views the EU and NATO with a lot of wariness, and has tried to stop them making inroads into former Soviet republics. This hasn't always worked - the Baltic states are now a part of the EU and NATO, and Georgia's bid to join NATO is now being fast-tracked.
On the 1st of March, Russian troops entered the Crimea, and quickly surrounded airports and government buildings in Sevastopol, Simferopol and Yalta. Currently, Ukrainian bases have being surrounded by Russian forces, as well as local militias loyal to the Kremlin. A break-away republic has been established, with its own Council of Ministers, a move that has been deemed unconstitutional by Kiev.
